```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
time = [0, 1, 2, 3, 4]
revenue = [200, 400, 650, 800, 850]
costs = [150, 500, 550, 550, 560]
```

```python
plt.plot(time, revenue)
plt.xlabel('Time')
plt.ylabel('Revenue')
plt.show()
```

```python
plt.plot(time, revenue, label='Revenue')
plt.plot(time, costs, label='Costs')
plt.xlabel('Time')
plt.ylabel('costs & Revenue')
plt.legend()
plt.show()
```

```python
#Task 2
plt.plot(time, revenue, label= 'Revenue', color ='purple',linestyle='--')
plt.xlabel('Time')
plt.ylabel('Revenue')
plt.show()
```

```python
plt.plot(time, revenue, label= 'Revenue', color ='purple',linestyle='--')
plt.plot(time, costs, label= 'Costs', color ='#82edc9',marker='s')
plt.xlabel('Time')
plt.ylabel('Revenue & Costs')
plt.legend()
plt.show()
```

```python
plt.plot(time, costs, label= 'Costs', color =(1,0,0,0.5),marker='D', alpha =0.2)
plt.xlabel('Time')
plt.ylabel('Costs')
plt.legend()
plt.show()
```

```python
#task 7
month_names= ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep","Oct", "Nov", "Dec"]
months = range(12)
conversion = [0.05, 0.08, 0.18, 0.28, 0.4, 0.66, 0.74, 0.78, 0.8, 0.81, 0.85, 0.85]
plt.xlabel("Months")
plt.ylabel("Conversion")
plt.plot(months, conversion)
plt.xticks(months, month_names)
plt.yticks([0.10,0.25,0.5,0.75],['10%','25%','50%','75%'])
plt.show()
```
