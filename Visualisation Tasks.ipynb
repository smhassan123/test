```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
time = [0, 1, 2, 3, 4]
revenue = [200, 400, 650, 800, 850]
costs = [150, 500, 550, 550, 560]
```

```python
plt.plot(time, revenue)
plt.xlabel('Time')
plt.ylabel('Revenue')
plt.show()
```

```python
plt.plot(time, revenue, label='Revenue')
plt.plot(time, costs, label='Costs')
plt.xlabel('Time')
plt.ylabel('costs & Revenue')
plt.legend()
plt.show()
```

```python
#Task 2
plt.plot(time, revenue, label= 'Revenue', color ='purple',linestyle='--')
plt.xlabel('Time')
plt.ylabel('Revenue')
plt.show()
```

```python
plt.plot(time, revenue, label= 'Revenue', color ='purple',linestyle='--')
plt.plot(time, costs, label= 'Costs', color ='#82edc9',marker='s')
plt.xlabel('Time')
plt.ylabel('Revenue & Costs')
plt.legend()
plt.show()
```

```python
plt.plot(time, costs, label= 'Costs', color =(1,0,0,0.5),marker='D', alpha =0.2)
plt.xlabel('Time')
plt.ylabel('Costs')
plt.legend()
plt.show()
```

```python
#task 5
months = range(12)
temperature = [36, 36, 39, 52, 61, 72, 77, 75, 68, 57, 48, 48]
flights_to_hawaii = [1200, 1300, 1100, 1450, 850, 750, 400, 450, 400, 860, 990, 1000]
```

```python
plt.subplot(1, 2, 1)
plt.plot(months, temperature, color='green')
plt.title('First Subplot')
plt.subplot(1,2,2)
plt.plot(flights_to_hawaii, temperature, color='blue')
plt.title('Second Subplot')
plt.show
```
